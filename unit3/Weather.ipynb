```python
import requests
import sqlite3 as lite
import time
from dateutil.parser import parse 
import collections
import pandas as pd
import datetime
```

```python
cities = { "Atlanta": '33.762909,-84.422675',
            "Austin": '30.303936,-97.754355',
            "Boston": '42.331960,-71.020173',
            "Chicago": '41.837551,-87.681844',
            "Cleveland": '41.478462,-81.679435'
        }
```

```python
con = lite.connect('weather.db')
cur = con.cursor()
with con:
    cur.execute('DROP TABLE weather')
    cur.execute('CREATE TABLE weather (city TEXT, time INT, temp NUMERIC )')
```

```python
for city in cities:
    utc_now = datetime.datetime.now() - datetime.timedelta(days=30)
    url = "https://api.forecast.io/forecast/113919b256fd6283e5236aee1b03afad/" + cities[city] + "," + utc_now.strftime('%s')
    print url
```

```python
for ciudad in cities:
    start_date = datetime.datetime.now() - datetime.timedelta(days=30)
    time = start_date
    for i in range(30):
        url = "https://api.forecast.io/forecast/113919b256fd6283e5236aee1b03afad/" + cities[ciudad] + "," + time.strftime('%s')
        r = requests.get(url)

        #a prepared SQL statement we're going to execute over and over again
        sql = "INSERT INTO weather (city,time,temp) VALUES (?,?,?)"

        city = r.json()['daily']['data'][0]
        #id, totalDocks, city, altitude, stAddress2, longitude, postalCode, testStation, stAddress1, stationName, landMark, latitude, location)
        cur.execute(sql,(ciudad,city['time'],city['apparentTemperatureMax']))
        time += datetime.timedelta(days=1)
```

```python
df = pd.read_sql_query("SELECT * FROM weather ORDER BY city",con,index_col='time')
```
