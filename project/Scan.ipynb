```python
dfTrain, dfTest = train_test_split(df, test_size=0.2)
```

```python
for k in range(1,5):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(dfTrain[:,7:17], dfTrain[:,17])
    # make predictions
    expected = dfTest[:,17]
    predicted = model.predict(dfTest[:,7:17])
    # misclassification rate
    error_rate = (predicted != expected).mean()
    print('%d:, %.2f' % (k, error_rate))
```

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
#import pydot 

X = df[['Vet Users','Staff Users','monthlyRxTot','clients','emails']]
y = df['bracket']

# fit a CART model to the data
model = DecisionTreeClassifier()
model.fit(X, y)
print(model)
#print model.predict([[8,5,20,10000,3000]])
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```
