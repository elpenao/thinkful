```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cm.csv', index_col=0)
df.dropna(inplace=true)
```

```python
%matplotlib inline
```

```python
data = df[['Average','Vet Users','Emails','Written Rx','Converted']]
```

```python
a = pd.scatter_matrix(data, alpha=0.05, figsize=(10,10), diagonal='hist')
```

```python
X = df[['Vet Users','Emails','Written Rx','Converted']]
y = df['Average']

## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

est.summary()
```
